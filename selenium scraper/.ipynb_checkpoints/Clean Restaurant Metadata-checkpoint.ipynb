```python
import numpy as np
import pandas as pd
import os
```

```python
df1 = pd.read_csv('Restaurant_data_filtered.csv')
df2 = pd.read_csv('Restaurant_MetaData.csv')
print(df1.head())
print(df2.head())
print(df1.shape)
print(df2.shape)
print(df1['name'].equals(df2['name']))
```

```python
columns = ['img_link', 'rating', 'n_ratings', 'address']
df1[columns] = df2[columns]
df1
```

```python
available_restaurants = [name[:-4] for name in os.listdir('reviews')]
print(available_restaurants[:5])
print(len(available_restaurants))
```

```python
df = pd.read_csv('Restaurant_metadata_full.csv')
print(df.shape)
```

```python
def filter_name(df):
    for restaurant in available_restaurants:
        if df['name'] == restaurant:
            return True
    return False

df = df[df.apply(filter_name, axis=1)]
print(df.shape)
```

```python
duplicates = df1[~df1['name'].duplicated(keep=False)]
duplicates
```

```python
duplicates[duplicates['name'].isin(available_restaurants)]['name'].nunique()
```
